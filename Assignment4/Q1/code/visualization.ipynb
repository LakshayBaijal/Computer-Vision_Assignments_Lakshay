```python
%cd /kaggle/working/Q1/Q1/code

from train import fit_model, TrainConfig
from vit import ViT
from cvt import CvT

# 1) Train ONE ViT (good config, 1D pos embedding)
vit = ViT(
    image_size=32,
    patch_size=4,
    dim=160,
    depth=5,
    heads=4,
    mlp_dim=256,
    pos_type="1D",
)
fit_model(
    vit,
    TrainConfig(model_name="vit_vis_only", epochs=20, lr=2e-4),
    output_dir="../training_curves/1.1_ViT/hyperparameter_tuning",
    aug_name="base",
)

# 2) Train ONE CvT standard
cvt = CvT(num_classes=10, use_pos_embed=False)
fit_model(
    cvt,
    TrainConfig(model_name="cvt_standard", epochs=20, lr=3e-4),
    output_dir="../training_curves/1.2_CvT/standard_vs_cvt",
    aug_name="base",
)
```
