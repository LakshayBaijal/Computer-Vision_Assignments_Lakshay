### Architecture discussion

The visual encoders are related but not identical in the task head and pooling details.
ImageNet RN50 ends in a 1000-way classifier (`fc`) optimized for supervised ImageNet labels.
CLIP RN50 visual encoder maps images into a shared embedding space and uses contrastive training against text embeddings.
CLIP additionally includes a separate text encoder (transformer), so the full CLIP model is multimodal.
