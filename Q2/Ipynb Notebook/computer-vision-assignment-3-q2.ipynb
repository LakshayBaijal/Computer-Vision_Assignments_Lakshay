```python
from kaggle_secrets import UserSecretsClient
import wandb
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Login
key = UserSecretsClient().get_secret("WANDB_API_KEY")
wandb.login(key=key)

entity = "mastermindlakshaybaijal-iiit-hyderabad"
project = "cv-assignment-q2"
runs = {
    "vanilla": "8fz0ch4k",
    "noskip": "47bslxn4",
    "residual": "napdthdd",
}

out_root = Path("/kaggle/working/wandb_exports_q2")
out_root.mkdir(parents=True, exist_ok=True)
api = wandb.Api(timeout=120)

plot_groups = {
    "losses": [
        "train/total_loss", "train/seg_loss", "train/depth_loss",
        "val/total_loss", "val/seg_loss", "val/depth_loss"
    ],
    "metrics": ["val/miou", "val/rmse"],
}

for name, run_id in runs.items():
    run = api.run(f"{entity}/{project}/{run_id}")
    rdir = out_root / name
    rdir.mkdir(parents=True, exist_ok=True)

    hist = run.history(samples=100000, pandas=True).reset_index(drop=True)
    hist.to_csv(rdir / "history.csv", index=False)
    pd.Series(dict(run.summary)).to_json(rdir / "summary.json", indent=2)
    pd.Series(dict(run.config)).to_json(rdir / "config.json", indent=2)

    # create plots safely
    for gname, keys in plot_groups.items():
        plt.figure(figsize=(10, 6))
        used = 0
        for k in keys:
            if k in hist.columns:
                s = pd.to_numeric(hist[k], errors="coerce").dropna()
                if len(s) > 0:
                    x = range(1, len(s) + 1)   # guarantees same length
                    plt.plot(x, s.values, label=k)
                    used += 1

        if used > 0:
            plt.title(f"{name} - {gname}")
            plt.xlabel("Logged points")
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.tight_layout()
            plt.savefig(rdir / f"{gname}.png", dpi=200)
        plt.close()

print("Saved:", out_root)
```

```python
from pathlib import Path
import json, shutil
import pandas as pd
import matplotlib.pyplot as plt

Q2_DIR = Path("/kaggle/working/Computer-Vision_Assignments_Lakshay/Q2")
OUT = Q2_DIR / "outputs"

# W&B exported histories root (adjust if needed)
WANDB_ROOT = Path("/kaggle/working/wandb_exports_q2")
if not WANDB_ROOT.exists():
    # try to auto-find nested export folder
    cands = list(Path("/kaggle/working").rglob("wandb_exports_q2"))
    if cands:
        WANDB_ROOT = cands[0]

FINAL = Path("/kaggle/working/final_submission/q2")
for d in ["vanialla_unet", "without_skip", "residual", "comparisons"]:
    (FINAL / d).mkdir(parents=True, exist_ok=True)

mapping = {
    "vanialla_unet": ("q2_multitask_unet_vanilla", "vanilla"),
    "without_skip": ("q2_multitask_unet_noskip", "noskip"),
    "residual": ("q2_multitask_unet_residual", "residual"),
}

rows = []
for dst_name, (run_dir_name, wb_name) in mapping.items():
    src = OUT / run_dir_name
    dst = FINAL / dst_name

    # 1) model + metrics
    for f in [src/"best.pt", src/"test_eval"/"test_metrics.json"]:
        if f.exists():
            shutil.copy2(f, dst/f.name)

    # 2) qualitative samples
    qsrc = src/"test_eval"/"qualitative_samples"
    if qsrc.exists():
        shutil.copytree(qsrc, dst/"qualitative_samples", dirs_exist_ok=True)

    # 3) history + graphs from W&B export
    hcsv = WANDB_ROOT / wb_name / "history.csv"
    if hcsv.exists():
        shutil.copy2(hcsv, dst/"history.csv")
        hist = pd.read_csv(hcsv)

        # losses plot
        plt.figure(figsize=(9,5))
        for k in ["train/total_loss","train/seg_loss","train/depth_loss","val/total_loss","val/seg_loss","val/depth_loss"]:
            if k in hist.columns:
                s = pd.to_numeric(hist[k], errors="coerce").dropna()
                if len(s): plt.plot(range(1,len(s)+1), s.values, label=k)
        plt.title(f"{wb_name} - Loss Curves")
        plt.xlabel("Epoch points"); plt.grid(alpha=0.3); plt.legend(); plt.tight_layout()
        plt.savefig(dst/"losses.png", dpi=180); plt.close()

        # metrics plot
        plt.figure(figsize=(9,5))
        for k in ["val/miou","val/rmse"]:
            if k in hist.columns:
                s = pd.to_numeric(hist[k], errors="coerce").dropna()
                if len(s): plt.plot(range(1,len(s)+1), s.values, label=k)
        plt.title(f"{wb_name} - Validation Metrics")
        plt.xlabel("Epoch points"); plt.grid(alpha=0.3); plt.legend(); plt.tight_layout()
        plt.savefig(dst/"metrics.png", dpi=180); plt.close()

    # read test metrics for comparison
    mfile = src/"test_eval"/"test_metrics.json"
    if mfile.exists():
        d = json.loads(mfile.read_text())
        rows.append({
            "model": wb_name,
            "test_miou": d.get("test_miou"),
            "test_rmse": d.get("test_rmse")
        })

# comparisons table + chart
cmp_dir = FINAL/"comparisons"
df = pd.DataFrame(rows).sort_values("model")
df.to_csv(cmp_dir/"metrics_table.csv", index=False)

md = ["| Model | Test mIoU | Test RMSE |","|---|---:|---:|"]
for _, r in df.iterrows():
    md.append(f"| {r['model']} | {r['test_miou']:.4f} | {r['test_rmse']:.4f} |")
(cmp_dir/"metrics_table.md").write_text("\n".join(md), encoding="utf-8")

plt.figure(figsize=(8,4))
x = range(len(df))
plt.bar([i-0.2 for i in x], df["test_miou"], width=0.4, label="mIoU")
plt.bar([i+0.2 for i in x], df["test_rmse"], width=0.4, label="RMSE")
plt.xticks(list(x), df["model"].tolist())
plt.title("Q2 Model Comparison (Test)")
plt.legend(); plt.grid(axis="y", alpha=0.3); plt.tight_layout()
plt.savefig(cmp_dir/"comparison_bar.png", dpi=180); plt.close()

# wandb links
(cmp_dir/"wandb_links.md").write_text(
"""Project: https://wandb.ai/mastermindlakshaybaijal-iiit-hyderabad/cv-assignment-q2
Vanilla: https://wandb.ai/mastermindlakshaybaijal-iiit-hyderabad/cv-assignment-q2/runs/8fz0ch4k
NoSkip: https://wandb.ai/mastermindlakshaybaijal-iiit-hyderabad/cv-assignment-q2/runs/47bslxn4
Residual: https://wandb.ai/mastermindlakshaybaijal-iiit-hyderabad/cv-assignment-q2/runs/napdthdd
""", encoding="utf-8")

# q2_report.md
report = f"""# Q2 Report: Multi-Task Learning (Segmentation + Depth)

## Test Metrics
{chr(10).join(md)}

## W&B
Project: https://wandb.ai/mastermindlakshaybaijal-iiit-hyderabad/cv-assignment-q2

## Comparison Summary
- Residual performs best overall (highest mIoU and lowest RMSE).
- NoSkip has slightly higher mIoU than Vanilla but much worse RMSE.
- Vanilla gives balanced depth quality but lower segmentation than Residual.

## Qualitative Discussion
- Compare object boundaries in masks across all three models.
- Compare depth smoothness and edge consistency.
- Include at least 10 side-by-side test predictions per model.
"""
(FINAL/"q2_report.md").write_text(report, encoding="utf-8")

print("Final folder ready at:", FINAL)
```
